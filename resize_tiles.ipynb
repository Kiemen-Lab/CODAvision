```python
def resize_tiles(path_tiles, target_size):
    imlist = [os.path.join(path_tiles, f) for f in os.listdir(path_tiles) if f.endswith('.png')]
    save_path = os.path.join(path_tiles, f'resized_{target_size}')
    os.makedirs(save_path, exist_ok=True)
    
    for i, img_path in enumerate(imlist):
        save_file_path = os.path.join(save_path, os.path.basename(img_path))
        if os.path.exists(save_file_path):
            # print(f'Image {os.path.basename(img_path)} exists already {i + 1}/{len(imlist)}')
            continue
        else:
            im = Image.open(img_path)
            im = im.resize((target_size, target_size), resample=Image.LANCZOS)
            print(f'Saving image: {os.path.basename(img_path)} {i + 1}/{len(imlist)}')
            im.save(save_file_path, format='PNG')
```

```python
target_size =1024
folders_1 = ['training','validation']
folders_2 = ['im','label']
for folder_1 in folders_1:
    path_1= os.path.join(r'\\10.99.68.52\Kiemendata\Valentina Matos\coda to python\test_png',folder_1)
    for folder_2 in folders_2:
        path_2 = os.path.join(path_1,folder_2)
        resize_tiles(path_2, target_size)
```

```python
import os
from PIL import Image
from glob import glob

def check_image_size(image_paths):
    for path in image_paths:
        try:
            with Image.open(path) as img:
                width, height = img.size
                if width != 1024 or height != 1024:
                    print(f"Image {path} has size {width}x{height} (expected 1024x1024)")
        except Exception as e:
            print(f"Error opening image {path}: {e}")

# Paths to training and validation datasets
pthDataset = r'\\10.99.68.52\Kiemendata\Valentina Matos\coda to python\test png'

# Define paths to training and validation directories
pthTrain = os.path.join(pthDataset, 'training')
pthValidation = os.path.join(pthDataset, 'validation')

# Get paths to training images and labels
Train_HE = sorted(glob(os.path.join(pthTrain, 'im', 'resized_1024', "*.png")))
Train_label = sorted(glob(os.path.join(pthTrain, 'label', 'resized_1024', "*.png")))

# Get paths to validation images and labels
Validation_HE = sorted(glob(os.path.join(pthValidation, 'im', 'resized_1024', "*.png")))
Validation_label = sorted(glob(os.path.join(pthValidation, 'label', 'resized_1024', "*.png")))

# Check image sizes
print("Checking training images...")
check_image_size(Train_HE)
print("Checking training labels...")
check_image_size(Train_label)
print("Checking validation images...")
check_image_size(Validation_HE)
print("Checking validation labels...")
check_image_size(Validation_label)
```

```python
import os

# For data preprocessing
from tensorflow import image as tf_image
from tensorflow import data as tf_data
from tensorflow import io as tf_io
from glob import glob
import tensorflow as tf

def check_imagedata(image_list, channels):
    corrupted_files = []
    for i, image_path in enumerate(image_list):
        print(f'Checking image: {i+1} of {len(image_list)}')
        try:
            
            # with tf.io.gfile.GFile(str(image_path), 'rb') as f:
            #     image = f.read()
            image = tf_io.read_file(image_path)
            image = tf.image.decode_png(image, channels=channels)
        except Exception as e:
            print(f"Error opening image {image_path}: {e}")
            corrupted_files.append(image_path)

    return corrupted_files

# Paths to training and validation datasets
pthDataset = r'C:\Users\Valentina\OneDrive - Johns Hopkins\Desktop\test png'
pthTrain = os.path.join(pthDataset, 'training')
pthValidation = os.path.join(pthDataset, 'validation')

# Get paths to training images and labels
Train_HE = sorted(glob(os.path.join(pthTrain, 'im', "*.png")))
Train_label = sorted(glob(os.path.join(pthTrain, 'label', "*.png")))

# Get paths to validation images and labels
Validation_HE = sorted(glob(os.path.join(pthValidation, 'im', "*.png")))
Validation_label = sorted(glob(os.path.join(pthValidation, 'label', "*.png")))

# Check for corrupted training images
corrupted_train_images = check_imagedata(Train_HE, channels=3)  # Color images
corrupted_train_labels = check_imagedata(Train_label, channels=1)  # Grayscale images

# Check for corrupted validation images
corrupted_validation_images = check_imagedata(Validation_HE, channels=3)  # Color images
corrupted_validation_labels = check_imagedata(Validation_label, channels=1)  # Grayscale images

print("Corrupted train images:", corrupted_train_images)
print("Corrupted train labels:", corrupted_train_labels)
print("Corrupted validation images:", corrupted_validation_images)
print("Corrupted validation labels:", corrupted_validation_labels)
            
```

```python
import os
from glob import glob

import numpy as np
from PIL import Image

def refactor_labels(image_list):
    corrected_labels = []
    for img_path in image_list:
        # Load the image
        img = np.array(Image.open(img_path))

        # Subtract 1 from the pixel values
        corrected_img = img - 1

        # Save the corrected image to a new path
        base_dir = os.path.dirname(img_path)
        save_dir = os.path.join(base_dir, 'corrected_labels')
        os.makedirs(save_dir, exist_ok=True)
        filename = os.path.basename(img_path)
        save_path = os.path.join(save_dir, filename)
        Image.fromarray(corrected_img).save(save_path)
        corrected_labels.append(save_path)

    return corrected_labels

# Paths to training and validation datasets
pthDataset = r'C:\Users\Valentina\OneDrive - Johns Hopkins\Desktop\python model 2\06_26_2024_raw_mattiles'
pthTrain = os.path.join(pthDataset, 'training')
pthValidation = os.path.join(pthDataset, 'validation')

# Get paths to training images and labels
Train_label = sorted(glob(os.path.join(pthTrain, 'label', "*.png")))

# Get paths to validation images and labels
Validation_label = sorted(glob(os.path.join(pthValidation,'label', "*.png")))

_ = refactor_labels(Train_label)
_ = refactor_labels(Validation_label)
```
