```python
from load_annotations import load_annotations
import os
import pickle
```

```python
#Function inputs
annotations_file = r'\\10.99.68.52\Kiemendata\Valentina Matos\coda to python\test model\data\SG_013_0061\annotations.pkl'
xmlfile = r'\\10.99.68.52\Kiemendata\Valentina Matos\coda to python\test model\SG_013_0061.xml'
dm='17-Apr-2024 16:56:26'
ra=None
```

```python
"""
    This script reads the xml file generated by the ImageScope Aperio system.
    
    First, it checks if annotation coordinates are already saved to the annotation
    data file. If not, it loads them for all images and saves them there.
"""

if ra is None:
    ra = 0
if dm is None:
    dm = []

print(' 1. of 4. Importing annotation data from xml file')

reduced_annotations, xyout_df = load_annotations(xmlfile)
reduced_annotations = float(reduced_annotations)
data_type=type(reduced_annotations)
print(data_type)
```

```python
if not xyout_df.empty:
    
    if ra == 1:
        reduced_annotations = ra
    xyout_df.iloc[:, 2:4] = xyout_df.iloc[:, 2:4] * reduced_annotations
    
    # Create the necessary directories if they don't exist
    annotations_dir = os.path.dirname(annotations_file)
    print(annotations_dir)
    if not os.path.exists(annotations_dir):
        os.makedirs(annotations_dir)
        
    if os.path.exists(annotations_file):
            print('File already exists, updating data...')
            with open(annotations_file, 'rb') as f:
                try:
                    existing_data = pickle.load(f)
                except EOFError:
                    existing_data = {}

            existing_data.update({'xyout': xyout_df.values, 'reduce_annotations': reduced_annotations, 'dm': dm})

            with open(annotations_file, 'wb') as f:
                pickle.dump(existing_data, f)
    else:
        print('Creating file...')
        with open(annotations_file, 'wb') as f:
            pickle.dump({'xyout': xyout_df.values, 'reduce_annotations': reduced_annotations, 'dm': dm}, f)

```
